### splitting dataset

```python
X = hr.drop(['Salary'], axis = 1)
y = hr['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = .2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
```

```python
from sklearn.linear_model import SGDRegressor, Ridge, LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeClassifier
```

```python
model_params = {'SVR':{'model':SVR(gamma = 'auto'),
                       'params':{ 'C':[.1,1,10,20,100],
                                 'kernel':['rbf','linear','poly'],
                                 
     }
  },
  'SGD':{
      'model':SGDRegressor(max_iter = 1000, tol = 1e-3),
      'params': {
          'eta0':[.001,.01,.1,.5,1,10,20,100]
      }

  },
  'Linear_regession':{
      'model':LinearRegression(),
      'params':{
          
      }
  },
  'Ridge_regression':{
      'model':Ridge(),
      'params':{
          'alpha':[0.05,0.01,.1,.5,1,10,100,200],
          'max_iter':[100,500,1000]
      }
  },
  'Decision_Tree':{
      'model':DecisionTreeClassifier(),
      'params':{
          'criterion':['gini','entropy'],
          'max_depth':[2,3,4,5]
      }
  }
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
scores = []
for model_name, model_parameter in model_params.items():
  clf = RandomizedSearchCV(model_parameter['model'],model_parameter['params'], cv = 2,
                           return_train_score = False, n_iter = 15)
  clf.fit(X_train,y_train)
  scores.append({
    'model':model_name,
    'best_score':clf.best_score_,
    'best_params':clf.best_params_
})
```
